```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from PIL import Image

from pathlib import Path

def load_images_from_folder(mask_folder, img_folder):
    masks = []
    images = []
    for filename in os.listdir(mask_folder):
        m = Image.open(os.path.join(mask_folder,filename))
        try:
            i = Image.open(os.path.join(img_folder, filename))
        except FileNotFoundError:
            print("The specified file does not exist.")
        if i is not None and m is not None:
            masks.append(m)
            images.append(i)
    return masks, images

mask_dir = "./videos/mask"
img_dir = "./videos"

masks, images = load_images_from_folder(mask_dir+"/log1", img_dir+"/log1")
print(len(masks))

logaddm, logaddi = load_images_from_folder(mask_dir+"/log2", img_dir+"/log2")
masks += logaddm
images += logaddi
print(len(masks))

logaddm, logaddi = load_images_from_folder(mask_dir+"/log3", img_dir+"/log3")
masks += logaddm
images += logaddi
print(len(masks))

logaddm, logaddi = load_images_from_folder(mask_dir+"/log4", img_dir+"/log4")
masks += logaddm
images += logaddi
print(len(masks))

masks = np.array(masks)
images = np.array(images)
plt.imshow(images[0])
plt.show()
plt.imshow(masks[0])
plt.show()

print(len(masks))
print(len(images))
```
